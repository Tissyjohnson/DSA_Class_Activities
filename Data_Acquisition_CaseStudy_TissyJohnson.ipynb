```python
import pandas as pd
```

```python
population=pd.read_csv("/content/population.csv")
population.head()
```

```python
gdp= pd.read_excel("/content/gdp.xlsx")
gdp.head()
```

```python
internet=pd.read_json("/content/internet_users.json")
internet.head()
```

```python
literacy=pd.read_xml("/content/literacy_rate.xml")
literacy.head()
```

```python
gdp["GDP"] = pd.to_numeric(gdp["GDP"], errors="coerce")
population["population"] = pd.to_numeric(population["population"], errors="coerce")
literacy["literacy_rate"] = pd.to_numeric(literacy["literacy_rate"], errors="coerce")
internet["internet_users"] = pd.to_numeric(internet["internet_users"], errors="coerce")
```

```python
population.isna().sum()
```

```python
gdp.isna().sum()
```

```python
internet.isna().sum()
```

```python
literacy.isna().sum()
```

```python
literacy=literacy.rename(columns={"name":"country"})
gdp=gdp.rename(columns={"Country":"country"})
```

```python
merged = (
    population.merge(gdp, on="country", how="outer")
              .merge(internet, on="country", how="outer")
              .merge(literacy, on="country", how="outer")
)
```

```python
merged
```

```python
merged["internet_penetration_rate"] = (merged["internet_users"] / merged["population"]) * 100
merged["gdp_per_capita"] = merged["GDP"] / merged["population"]
```

```python
merged
```

```python
merged.isna().sum()
```

```python
duplicates = merged[merged.duplicated()]
print(duplicates)
```

```python
print( merged[(merged["population"] < 0)| (merged["GDP"] < 0) | (merged["internet_users"] < 0) ])
```

```python
print(merged.isnull().sum())
```

```python
top_internet = merged[["country", "internet_penetration_rate"]] \
    .sort_values(by="internet_penetration_rate", ascending=False)

print(top_internet.head(10))
```

```python
avg_literacy = merged["literacy_rate"].mean()
print("Average literacy rate across countries:", avg_literacy)
```

```python
corr = merged[["GDP", "population", "literacy_rate", "internet_users"]].corr()
print(corr)
```

```python
import sqlite3
```

```python
conn = sqlite3.connect("wholedata.db")
```

```python
merged.to_sql("countries", conn, if_exists="replace", index=False)
```

```python
cursor = conn.cursor()
```

```python
cursor.execute("""SELECT country, internet_penetration_rate
FROM countries
ORDER BY internet_penetration_rate DESC
LIMIT 10;"""
)
```

```python
high=cursor.fetchall()
for i in high:
  print(i)
```

```python
df2= pd.read_sql_query("""SELECT AVG(literacy_rate) AS avg_literacy_rate
FROM countries;""",conn)
print(df2.to_string(index=False))
```

```python
df_corr = pd.read_sql_query("""
SELECT literacy_rate, internet_penetration_rate
FROM countries;
""", conn)

print("Correlation Literacy - Internet Penetration:",
      df_corr["literacy_rate"].corr(df_corr["internet_penetration_rate"]))
```

```python
cursor.execute("SELECT * FROM countries LIMIT 5")
rows=cursor.fetchall()
for row in rows:
  print(row)

```

```python
cursor = conn.execute("""
SELECT country, gdp_per_capita
FROM countries
ORDER BY gdp_per_capita DESC;
""")

for row in cursor.fetchall():
    print(row)
```

```python
df_gdp = pd.read_sql_query("""
SELECT country, gdp_per_capita
FROM countries
WHERE gdp_per_capita > 10000
ORDER BY gdp_per_capita DESC;
""", conn)

print(df_gdp.to_string(index=False))
```

```python
df_total_pop = pd.read_sql_query("""
SELECT SUM(population) AS total_population
FROM countries;
""", conn)

print(df_total_pop.to_string(index=False))
```

```python
df_low_lit = pd.read_sql_query("""
SELECT country, literacy_rate, internet_penetration_rate
FROM countries
ORDER BY literacy_rate ASC
LIMIT 10;
""", conn)

print(df_low_lit.to_string(index=False))
```

```python
df_top_gdp = pd.read_sql_query("""
SELECT country, GDP, population, gdp_per_capita
FROM countries
ORDER BY GDP DESC
LIMIT 5;
""", conn)

print(df_top_gdp.to_string(index=False))
```

```python
df_internet = pd.read_sql_query("""
SELECT country, internet_users, population, internet_penetration_rate
FROM countries
WHERE internet_penetration_rate > 70
ORDER BY internet_penetration_rate DESC;
""", conn)

print(df_internet.to_string(index=False))
```

```python
df_avg_gdp = pd.read_sql_query("""
SELECT AVG(gdp_per_capita) AS avg_gdp_per_capita
FROM countries
WHERE internet_penetration_rate > 50;
""", conn)

print(df_avg_gdp.to_string(index=False))
```

```python
df_lit = pd.read_sql_query("""
SELECT COUNT(*) AS num_countries, AVG(internet_penetration_rate) AS avg_internet_penetration
FROM countries
WHERE literacy_rate > 90;
""", conn)

print(df_lit.to_string(index=False))
```
