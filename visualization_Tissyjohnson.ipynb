```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
store = pd.read_excel("/content/Sample - Superstore.xlsx",sheet_name='Orders')
store
store.columns = store.columns.str.strip()

print(store.head())
print(store.columns.tolist())
```

```python
# Histogram for Sales
plt.figure(figsize=(8,5))
sns.histplot(store['Sales'], bins=50, kde=True)
plt.title("Distribution of Sales")
plt.show()



```

```python
# Histogram for Profit
plt.figure(figsize=(8,5))
sns.histplot(store['Profit'], bins=50, kde=True, color="orange")
plt.title("Distribution of Profit")
plt.show()
```

```python
# Countplot for Category
plt.figure(figsize=(6,4))
sns.countplot(x="Category", data=store, palette="Set2")
plt.title("Count of Orders by Category")
plt.show()
```

```python
# Sales by Category
plt.figure(figsize=(5,4))
sns.barplot(x="Category", y="Sales", data=store, estimator=sum)
plt.title("Total Sales by Category")
plt.show()
```

```python
# Sales vs Profit
plt.figure(figsize=(8,6))
sns.scatterplot(x="Sales", y="Profit", data=store, alpha=0.6)
plt.title("Sales vs Profit")
plt.show()
```

```python
# Category vs Sales
plt.figure(figsize=(8,6))
sns.boxplot(x="Category", y="Sales", data=store)
plt.title("Sales Distribution by Category")
plt.show()
```

```python
# Region vs Profit
plt.figure(figsize=(8,6))
sns.barplot(x="Region", y="Profit", data=store, estimator=sum)
plt.title("Total Profit by Region")
plt.show()
```

```python
# Sales vs Profit by Category
plt.figure(figsize=(8,6))
sns.scatterplot(x="Sales", y="Profit", hue="Category", data=store, alpha=0.7)
plt.title("Sales vs Profit by Category")
plt.show()
```

```python
# Sales by Region & Category
plt.figure(figsize=(10,6))
sns.barplot(x="Region", y="Sales", hue="Category", data=store, estimator=sum)
plt.title("Total Sales by Region and Category")
plt.show()
```

```python
# Pairplot for numeric variables
sns.pairplot(store[["Sales","Profit","Quantity","Discount"]], diag_kind="kde")
plt.suptitle("Multivariate Relationships", y=1.02)
plt.show()
```

```python
# Top 10 Sub-Categories by Sales
plt.figure(figsize=(5,4))
subcat_sales = store.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(10)
sns.barplot(x=subcat_sales.values, y=subcat_sales.index)
plt.title("Top 10 Sub-Categories by Sales")
plt.show()
```
