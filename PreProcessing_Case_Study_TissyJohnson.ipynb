```python
missing = house.isnull().sum()


# Impute missing values
for col in missing[missing > 0].index:
    if house[col].dtype in ['int64', 'float64']:   # Numerical column
        median_value = house[col].median()
        house.loc[:,col]=house[col].fillna(median_value)
        print(f"{col}: Median imputation (value = {median_value})")
```

Handled Missing Value Columns [Sale Price , No of Bedrooms ,  No of Bathrooms , Flat Area (in Sqft) , Lot Area (in Sqft) , Area of the House from Basement (in Sqft) ,  Zipcode , Latitude , Longitude , Living Area after Renovation (in Sqft)] with Median of Respective Columns.

```python
minmax_cols = ["Flat Area (in Sqft)", "Lot Area (in Sqft)",
               "Area of the House from Basement (in Sqft)",
               "Basement Area (in Sqft)",
               "Living Area after Renovation (in Sqft)",
               "No of Floors"]
mscaler = MinMaxScaler()
house[minmax_cols] = mscaler.fit_transform(house[minmax_cols])

standard_cols = ["No of Bedrooms", "No of Bathrooms",
                 "Overall Grade", "Age of House (in Years)",
                 "Latitude", "Longitude"]

sscaler = StandardScaler()
house[standard_cols] = sscaler.fit_transform(house[standard_cols])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
houselabel= LabelEncoder()
house['Waterfront View']= houselabel.fit_transform(house['Waterfront View'])
house['Condition of the House']= houselabel.fit_transform(house['Condition of the House'])
```

```python
house["Year_Sold"] = house["Date House was Sold"].dt.year
house["Month_Sold"] = house["Date House was Sold"].dt.month
house["Day_Sold"] = house["Date House was Sold"].dt.day
house=house.drop('Date House was Sold',axis = 1)
```

```python
house=house.drop("ID",axis=1)
```

```python
num_cols = house.select_dtypes(include=['int64', 'float64']).columns.tolist()

if "Sale Price" in num_cols:
    num_cols.remove("Sale Price")
```

```python
for column in num_cols:
    Q1 = house[column].quantile(0.25)
    Q3 = house[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    house[column] = house[column].clip(lower=lower_bound, upper=upper_bound)
```

```python
import numpy as np
house['Sale Price'] = np.log1p(house['Sale Price'])
```
